```python
import pandas as pd

df = pd.read_csv("/content/sample_data/student_dataset.csv")
print(df.head(5))
```

```python
print(df.columns)
```

```python
print(df.dtypes)
```

```python
print(df.shape)
```

```python
average_marks = df['Student Marks'].mean()
print(average_marks)
```

```python
hiest_marks = df['Student Marks'].max()
print(hiest_marks)
```

```python
lowes_marks = df["Student Marks"].min()
print(lowes_marks)
```

```python
hiest_score = df[df["Student Marks"]>70]
print(hiest_score)
```

```python
filtered_student = df[(df['Study Hours']>3) & (df['Student Marks'] < 70)]
print(filtered_student)
```

```python
import numpy as np
df['Result'] = np.where(df['Student Marks'] >= 50 , 'Pass', 'Fail')

df.head()
```

```python
hiest_Student = df.loc[df['Student Marks'].idxmax()]
print(hiest_Student)
```

```python
lowest_Student = df.loc[df['Student Marks'].idxmin()]
print(lowest_Student)
```

```python
df_sorted = df.sort_values(by='Student Marks', ascending=False)
print(df_sorted)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(df['Student Name'], df['Student Marks'], color='skyblue')

plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.title('Student Marks Distribution chart')


plt.xticks(rotation=45)


plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Study Hours'], df['Student Marks'], color='purple', alpha=0.7)

plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Relationship Between Study Hours and Marks')

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='Study Hours')

plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Study Hours'], df_sorted['Student Marks'], marker='o', color='red', linestyle='-', linewidth=2)


plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Trend: Study Hours vs. Marks')


plt.grid(True, linestyle='--', alpha=0.5)


plt.show()
```
